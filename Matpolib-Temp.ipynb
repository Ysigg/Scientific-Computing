```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
years = np.array([1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999])
temperatures = np.array([14.1999, 14.2411, 14.0342, 14.2696, 14.197, 14.3055, 14.1853, 14.3577, 14.4187, 14.3438])

# Estimated temperature in even years using linear interpolation
even_years = np.arange(1982, 2000, 2)
linear_interpolation = np.interp(even_years, years, temperatures)

# Least-square regression
coefficients = np.polyfit(years, temperatures, deg=1)
regression = np.polyval(coefficients, even_years)

# Plotting the data and curves
plt.plot(years, temperatures, 'bo', label='Data')  # Data points
plt.plot(even_years, linear_interpolation, 'r-', label='Linear Interpolation')  # Linear interpolation
plt.plot(even_years, regression, 'g-', label='Least-Square Regression')  # Regression curve

plt.xlabel('Year')
plt.ylabel('Temperature (C)')
plt.title('Relationship between Temperature and Year')
plt.legend()

plt.show()
```
